```python
# Normalizing the Scale, where Min = 0, Max = 1

def norm_fun(i):
    x = (i - i.min())/(i.max() - i.min())
    return x
```

```python
data_norm = norm_fun(data)
```

```python
data_norm.describe()
```

```python
# Performing KMeans Clustering

from sklearn.cluster import KMeans
```

```python
TWSS = []
k = list(range(2,9))

for i in k:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(data_norm)
    TWSS.append(kmeans.inertia_)
    
TWSS
```

```python
plt.plot(k, TWSS, 'ro-')
plt.xlabel('Num of Clusters')
plt.ylabel('Total With in SS')
plt.show()
```

```python
list = []
for k in range(2,10):
    kmeans = KMeans(n_clusters = k, init = 'random', random_state = 1, max_iter = 30, n_init = 10)
    kmeans.fit(data_norm)
    list.append(kmeans.inertia_)
    
# Performing KMeans Clustering   
    
from kneed import KneeLocator
k1 = KneeLocator(range(2,10), list , curve = 'convex')
k1.elbow

plt.style.use("seaborn")
plt.plot(range(2,10), list)
plt.xticks(range(2,10))
plt.ylabel("Inertia")
plt.axvline(x= k1.elbow, color = 'r', label = 'axvline - full height', ls = '--')
plt.show()
```
